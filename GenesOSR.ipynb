```python
def build_ffnn(input_shape, loss, opt):

    # Build the neural network layer by layer
    input_layer = tfkl.Input(shape=input_shape, name='Input')
    hidden_layer1 = tfkl.Dense(units=32, activation='relu', name='Hidden1')(input_layer)
    hidden_layer2 = tfkl.Dense(units=16, activation='relu', name='Hidden2')(hidden_layer1)
    output_layer = tfkl.Dense(units=3, activation='softmax', name='Output')(hidden_layer2)

    # Connect input and output through the Model class
    model = tf.keras.Model(inputs=input_layer, outputs=output_layer, name='FFNN')

    # Compile the model
    model.compile(loss=loss, optimizer=opt, metrics='accuracy')

    # Return the model
    return model
```
